```python
import pandas as pd
```

```python
data = pd.read_csv('full_thuat_ngu.csv')
```

```python
data
```

```python
data = data[['thuatngu','mota', 'nguon', 'link', 'tinh']]
```

```python
data
```

```python
data['thuatngu'][0]
```

```python
def processing_data(thuatngu):
    if 'Hết hiệu lực' in thuatngu:
        return 1
    return 0
```

```python
data['label'] = data['thuatngu'].map(lambda x : processing_data(x))
```

```python
data
```

```python
data['thuatngu'][3]
```

```python
def drop_hethieuluc(thuatngu):
    if 'Hết hiệu lực' in thuatngu:
        return thuatngu.replace('\nHết hiệu lực', '').replace('.', '')
    return thuatngu
```

```python
data['thuatngu'] = data['thuatngu'].map(lambda x : drop_hethieuluc(x))
```

```python
data
```

```python
def drop_dotc(thuatngu):
    if '.' in thuatngu:
        return thuatngu.replace('.', '')
    return thuatngu
```

```python
data['thuatngu'] = data['thuatngu'].map(lambda x : drop_dotc(x))
```

```python
data
```

```python
data['mota'] = data['mota'].map(lambda x : (x.lower()))
```
